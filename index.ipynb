Now, let's try to find out **how many countries participated in the 1986 World Cup**.

Hint 1: As a first step, create a new dataset that only contains games in that year.

Hint 2: Make sure you don't end up with duplicate country names. Consider using `set()` or `.unique()`.

```python
# Your code here
# filter the year required.
participated_countries = len(set(df.loc[df["Year"] == 1986 , ["Home Team Name" , "Away Team Name"]].values.flatten()))
#.values.flatten() combines both columns into a single 1D array so that we can extract all team names from both sides.
participated_countries
```

```python
#Another learned concept of complex filtering using loc[]

# Using .loc with combined conditions (grouped properly)
unique_teams = set(df.loc[(df["Year"] == 1986) & ((df["Home Team Name"].notnull()) | (df["Away Team Name"].notnull())), 
                          ["Home Team Name", "Away Team Name"]].values.flatten())

len(unique_teams)
```

In World Cup history, **how many matches had 5 goals or more in total**? Create a column `"Total Goals"` to answer this question.

```python
# Your code here
# create a new column with the sum of the matches in both team names that have 5\more goals.
df["Total Goals"] = df["Home Team Goals"] + df["Away Team Goals"]
#filter the data of thedf["Total Goals"] to matches more ornwithin 5 goals
df.loc[df["Total Goals"]>=5]
# check the length of the data
len(df.loc[df["Total Goals"]>=5])
```

Run the code below. You'll notice that for Korea, there are records for both North-Korea (Korea DPR) and South-Korea (Korea Republic). 

```python
# Run this cell without changes

# Display all records containing the string 'Korea'
df.loc[df['Home Team Name'].str.contains('Korea'), 'Home Team Name']
```

Imagine that, for some reason, we simply want Korea listed as one entry, so we want to replace every "Home Team Name" and "Away Team Name" entry that contains "Korea" to simply "Korea". In the same way, we want to change the columns "Home Team Initials" and "Away Team Initials" to NSK (North & South Korea) instead of "KOR" and "PRK". 

```python
#update the "Home Team Name" and the "Home Team Initials columns
# Replace team names containing "Korea" with "Korea"
df["Home Team Name"] = df["Home Team Name"].apply(lambda x: "Korea" if "Korea" in x else x)
df["Away Team Name"] = df["Away Team Name"].apply(lambda x: "Korea" if "Korea" in x else x)

# Replace "KOR" and "PRK" initials with "NSK"
df["Home Team Initials"] = df["Home Team Initials"].replace({"KOR": "NSK", "PRK": "NSK"})
df["Away Team Initials"] = df["Away Team Initials"].replace({"KOR": "NSK", "PRK": "NSK"})
df.update(df["Home Team Name"].apply(lambda x: "Korea" if "Korea" in x else x))
df.update( df["Home Team Initials"].replace({"KOR": "NSK", "PRK": "NSK"}))
```
